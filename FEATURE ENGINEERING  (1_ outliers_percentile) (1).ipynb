```python
import pandas as pd
```

```python
df = pd.read_csv('Downloads/weight-height.csv')
df
```

## Detect outliers using percentile

```python
max_thresold = df['Height'].quantile(0.85)
max_thresold
```

```python
df[df['Height']>max_thresold]
```

```python
min_thresold = df['Height'].quantile(0.15)
min_thresold
```

```python
df[df['Height']<min_thresold]
```

## Remove outliers

```python
df[(df['Height']<max_thresold) & (df['Height']>min_thresold)]
```
